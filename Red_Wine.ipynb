```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
#visualization
df.quality.value_counts().plot(kind='bar', color='skyblue'
, figsize=(8, 5))
plt.title('Wine Quality Distribution')
plt.xlabel('Quality')
plt.ylabel('Count')
```

```python
df.head()
```

```python
for column in df.columns:
    sns.histplot(df[column],kde=True)
```

```python
# Example: Distribution of Alcohol
sns.histplot(df['alcohol'], kde=True, color='green')
plt.title("Distribution of Alcohol")
plt.xlabel("Alcohol %")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x=df['alcohol'], color='orange')
plt.title("Boxplot of Alcohol")
plt.show()
```

```python

# Use scatterplot to visualize alcohol content across wine quality
sns.scatterplot(x='quality', y='alcohol', data=df)
plt.title("Alcohol Content Across Wine Quality")
plt.show()
```

```python
# Pick only important columns
selected_cols = ['alcohol', 'sulphates', 'volatile acidity', 'citric acid', 'quality']

# Create pair plot
sns.pairplot(df[selected_cols], hue='quality', palette='Set1')
plt.suptitle("Pair Plot of Important Wine Features", y=1.02)
plt.show()
```

```python
sns.catplot(x='quality', y='alcohol', data=df, kind='box', palette='Set2')
```
