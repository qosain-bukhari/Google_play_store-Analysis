```python
#feature engineering process
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['year']=df['Date_of_Journey'].str.split('/').str[2]
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['year']=df['year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python

df['Hour']=df['Dep_Time'].str.split(':').str[0]
df['Minutes']=df['Dep_Time'].str.split(':').str[1]
df['Hour']=df['Hour'].astype(int)
df['Minutes']=df['Minutes'].astype(int)
```

```python

df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python

df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python

df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes']=df['Arrival_Time'].str.split(':').str[1]
```

```python

df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Minutes']=df['Arrival_Minutes'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python

df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].unique()
```

```python
# Create a dictionary to map values
stops_map = {
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
}

# Apply the mapping
df['Total_Stops'] = df['Total_Stops'].map(stops_map)
```

```python
df['Total_Stops'].isnull().sum()
```

```python
# Fill missing with most common value
df['Total_Stops'] = df['Total_Stops'].fillna(1)  
```

```python
df["Total_Stops"].head()
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Duration_Hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Hours']
```

```python
df['Airline'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oneHotEncoder=OneHotEncoder()
```

```python
Encoded_Airline=oneHotEncoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
Encoded_Airline
```

```python
Encoded_df=pd.DataFrame(Encoded_Airline,columns=oneHotEncoder.get_feature_names_out(['Airline','Source','Destination']))
```

```python
df = pd.concat([df, Encoded_df], axis=1)
df
```

```python
import matplotlib.pyplot as plt
sns.histplot(df['Price'],kde=True,color='blue' )
plt.title('Price Distribution')
plt.xlabel('Price') 
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='Price', data=df)
plt.xticks(rotation=90)
plt.title("Price Range by Airline")
plt.xlabel("Airline")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='Total_Stops', y='Price', data=df)
plt.title("Stops vs Average Flight Price")
plt.xlabel("Total Stops")
plt.ylabel("Average Price")
plt.show()
```
